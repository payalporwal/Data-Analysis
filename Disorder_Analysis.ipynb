# **Data Visualization**

It begins with plotting a graph to see overall which disorder is standing where in terms of rate of change.

```python
sns.set_style("whitegrid")
plt.figure(figsize=(10,6))
data_wide = data.pivot('Year','Entity')
sns.lineplot(data=data_wide)
```

Rate of Anxiety Disorder is highest while Schizophrenia is the least. 
And the rest are in between them. 

Anxiety and Depression is very high with respect to others. 
While, Eating, Alcohol and Drug usage are at almost similar rates.

Schizophrenia and Bipolar Disorder are at the least.

```python
plt.figure(figsize=(20,8))
pivot = data.pivot_table(
    index=["Year"],
    columns = ['Entity'],
    values = disorders
)
sns.heatmap(pivot,linecolor='white')
```

Here, We can see that Anxiety and Depression are more prominent than other disorder. We can also observe slight higher rate for Alcohol Use and Drug use Disorder as well.

Also, Countries showing high rates of Anxiety are Canada, Eygpt, Iceland, Israel, New Zealand and few more.

While, high Depression rates are in countries like Bangladesh, Indonesia, Japan, North Korea, Russia and few more.

The brightest line can be observed for Anxiety Disorder at New Zealand.

```python
disorders =[ 'Schizophrenia (%)', 'Bipolar disorder (%)',
       'Eating disorders (%)', 'Anxiety disorders (%)',
       'Drug use disorders (%)', 'Depression (%)',
       'Alcohol use disorders (%)']
```

```python
plt.figure(figsize=(10,8))
color = ['r', 'g', 'b', 'orange']
max_values =data.groupby("Entity").max()['Anxiety disorders (%)'].sort_values(ascending=False).head(10)
square= squarify.plot(sizes=max_values.values,label=max_values.index, value=max_values.values, color=color,alpha=.7)
plt.title('Anxiety Rate across Top Countries')
```

New Zealand has the highest Anxiety Rate. The same observation we had in the heatmap.

```python
sns.set_style("darkgrid")
g=sns.PairGrid(data,x_vars=['Entity'],y_vars=disorders,hue="Year",height=5, aspect=7)
g.map(sns.lineplot,color=".3")
```



Bipolar Disorders graph is not deviated much in the range of these Years(1990-2017). Though, Schizophrenia is showing comparatively more deviation.

Eating Disorders have deviated a lot and consistently in all the countries. We can see it in Alcohol and Drug use Disorders as well. This must be because of change in lifestyle within these years.

We can refer that these changes leads to Depression and Anxiety Disorders, for sure.






```python
sns.set_style("whitegrid")
plt.figure(figsize=(10,6))
India_graph = data.query('Entity=="India"').drop(columns=['Year'],axis=1)
sns.lineplot(data=India_graph)
```

In India, Depression Rate is higher rather than Anxiety Rate, while Eating Disorder is the least rather than Schizophrenia

From the above graph, first, we can see if there is any linkage between increasing rate of Anxiety and Depression, with increasing bad eating habit , alcohol and drug abuse.
Second, we can think of though Schizophrenia and Bipolar Disorder are genetically transmitted, can bad lifestyle and eating habits increase the their risk factor as well?

```python
sns.set_style("dark")
g = sns.PairGrid(data[disorders])
g.map_diag(sns.histplot)
g.map_offdiag(sns.regplot)
```

We can notice the positive inclination, which refers to impact of one disorder to the other.

And the answer to both the above observation is yes.





Finally, the conclusion we get :




*   There is a increasing in bad eating habits, alcohol and drugs consumption and bad lifestyle in these years.
*   This bad lifestyle is impacting mental disorders and higher risk of occurrence among people.

*   Even, those disorder which were genetically transmitted, eating habits are impacting rate of occurrence as well.
*   Drug consumption is more likely to impact than alcohol consumption.
